##WMP Bloodfeed Efficiency Analysis: Summary and Scope

This analysis focuses on intepreting efficiency trends, identifying data limitations, and outline process improvement recommendations with the context of blood feeding efficiency for WMP's MMFPA facilities. Data has been gathered from over a three month period (1/09/2025 - 19/12/2025) analysing the relationship between blood input and egg output within production TML and LAO lines. 

##Dataset Overview
Number of trials - 158

Key variables analysed:
- Input quantity (blood_input_ml)
- Output quantity (egg_output_total)

Data processed and validated within notebook 1.

Data patterns were explored via notebooks 2, 3, and 4



##Key Findings and Intepretation
Findings indicate that peak efficiency (eggs per ml) is reached around 2400ml of blood over a two week feeding period (4 feeds per week, separated over 48hr periods) at 758% efficiency. This peak appears to slope off at higher blood inputs and reduce to 577% and 595%.

The peak efficiency rate can be considered as reliable due the vast majority of trials being performed were done at the 2400ml interval (130 out of 158 total). Lower efficiencies discovered from lower inputs (73% at 0-500ml, 486% at 500-1000ml, 701% at 1500-2000ml) makes logical sense in regards to the discovered peak efficiency. However and most interestingly, a slight uptick in efficiency was discovered at the 3000-3500ml input mark from the 2500-3000ml input mark, being 596% and 577% respectively. 

Further analysis was conducted regarding peak efficiency by line, breaking up trials into TML AMP, TML BRO, LAO BRO, and LAO AMP groups. Results of this analysis revealed that LAO BRO and LAO AMP groups had roughly equivalent efficiencies, while TML AMP and TML BRO had the largest difference in efficiencies (TML AMP lowest of all groups, TML BRO as highest).

##Limitations
It is important to highlight that all efficiency results outside of the ones denoting peak efficiency are not neccessarily reliable for drawing conclusions and making predictions. This is due to the incredibly low number of trials outside of those that ran at 2400ml of blood input.

Similar limitations exist for investigations of efficiencies by line, with the bulk of trials being performed in LAO BRO and TML BRO and the fewest occuring in TML AMP. 

This lack of trial data variance is important because the variance found in trials with 2400ml blood input covers the range of results in all other blood input brackets, perhaps suggesting that the variance in these other groups could be just as large given enough trials. 

##Recommendations

It is recommended that a systemic analysis be conducted to more conclusively compare efficiency rates at different concentrations and blood inputs. This would allow for confidence in identifying peak efficiency. If this is not possible, then at minimum it is recommended that blood input above 2400ml be further investigated. The slight uptick discovered in the 3500ml bracket indictates that peak efficiency could be achieved at higher blood inputs.

##Final Summary

Human and Horse blood acquisition represents a considerable, necessary cost to MMPFA operations, being crucial to egg production. The proper establishment of peak efficiency in terms of blood input to egg output is critical to limiting costs and reducing wastage. Excess blood could be better allocated to underfed cages, saved for later or not ordered at all. At present, peak efficiency at current cage and population sizes can be considered to be 2400ml over a two week feeding period (4 feeds a week). This amount has a strong amount of trial data supporting it, but a lack of data around other amounts could be obscuring the true peak or drop off in efficiency from increased feeding.
